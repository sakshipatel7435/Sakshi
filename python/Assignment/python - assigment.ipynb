Problem Statement: Password Generator

Make a program to generate a strong password using the input given by the user. To generate a password, randomly take some words from the user input and then include numbers, special characters and capital letters to generate the password. Also, keep a check that password length is more than 8 characters.

Note: Include Exception handling wherever required. Also, make a 'User' class and store the details like user id, name and password of each user as a tuple.

---

```python
import random
import string

class User:
    def __init__(self, user_id, name, password):
        self.details = (user_id, name, password)

def generate_password(user_words):
    all_chars = string.ascii_uppercase + string.digits + "!@#$%&*"
    password = ""

    for i in range(3):
        password += random.choice(user_words)

    for i in range(5):
         password += random.choice(all_chars)

    random_password = ''.join(random.sample(password, len(password)))

    if len(random_password) < 8:
        random_password += random.choice(all_chars)

    return random_password

words = ["hello", "python", "user", "code"]
pwd = generate_password(words)
u1 = User(1,"sakshi", pwd)

print("Generated Password:", pwd)
print("User Details:", u1.details)
```

17.Write a python program using function to find the sum of odd series andevenseries
Odd series: 12/ 1! +32/ 3! + 52/ 5!+……n
Even series: 22/ 2! + 42/ 4! + 62/ 6!+……n

```python
import math  

# Odd series sum
def odd_series_sum(n):
    total = 0
    num = 1
    for i in range(n):
        total += (num ** 2) / math.factorial(num)
        num += 2
    return total

# Even series sum
def even_series_sum(n):
    total = 0
    num = 2
    for i in range(n):
        total += (num ** 2) / math.factorial(num)
        num += 2
    return total

# User input
terms = int(input("term?: "))

# Output
print("Odd series no sum:", odd_series_sum(terms))
print("Even series no sum:", even_series_sum(terms))

```

```python
import math

def sum_of_series(n):
    odd_sum = 0
    even_sum = 0

    for i in range(1, n + 1):
        if i % 2 == 1:
            odd_sum += i**2 / math.factorial(i)
        else:
            even_sum += i**2 / math.factorial(i)

    print("Odd series sum =", odd_sum)
    print("Even series sum =", even_sum)

sum_of_series(6)
```

Problem Statement : Password Generator
Make a program to generate a strong password using the input given by theuser. To generate a password, randomly take some words from the user input andtheninclude numbers, special characters and capital letters to generate the password. Also, keep a check that password length is more than 8 characters. Note: Include Exception handling wherever required. Also, make a ‘User’ classand store the details like user id, name and password of each user as a tuple.

```python
import random

user_input = input("your password: ")
# maximum 4 characters j allow
user_input = user_input[:4]
words = user_input.split()

if len(words) >= 2:
    chosen = random.sample(words, 2)
else:
    chosen = words

new_words = [word.capitalize() for word in chosen]

base = ""
for word in new_words:
    base = base + word

numbers = str(random.randint(100, 999))       
symbols = random.choice(['@', '#', '$', '%'])

password = base + numbers + symbols

while len(password) < 8:
    password += random.choice('abcxyz123')
    
print("your storng password: ", password)
```
