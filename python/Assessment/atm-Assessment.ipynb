```python
class atm:
    def __init__(self):
        self.pin = ""
        self.balance = 0
        self.name = ""
        self.no = 0
        self.transcations = []
        self.card_activation()



    def menu(self):
        user_input = input(
            """
        1. Press 1 to change the pin
        2. Press 2 to deposit money
        3. Press 3 to withdraw money
        4. Press 4 to check balance
        5. Press 5 to view transaction history
        6. Press 6 to exit
        """
        )
        if user_input == "1":
            self.change_pin()
        elif user_input == "2":
            self.deposit_money()
        elif user_input == "3":
            self.withdraw_money()
        elif user_input == "4":
            self.check_balance()
        elif user_input == "5":
            self.transaction_history()
        else:
            exit()



    def create_pin(self):  # 4 digit int number
        print("Hello, Welcome to ATM system")
        user_pin = input("enter the 4 digit to set it as your pin")
        self.pin = user_pin
        if len(self.pin) == 4 and self.pin.isdigit():
            print("Your pin has been created successfully")
            self.deposit_money()
        else:
            print("Invalid PIN! Please enter exactly 4 digits.")
            

    def card_activation(self):
            name = input("Enter your name")
            self.name = name
            no = input("Enter you mobile number")
            self.no = no
            if len(self.no) == 10 and self.no.isdigit():
                print("Your card has been activated successfully")
                self.create_pin()
            else:
                print("Invalid no! Please enter exactly 10 digits.")
                self.card_activation()

    def deposit_money(self):
        user_pin = input("enter the pin to deposit your money")  # min 500 rs deposit is necessary
        if user_pin == self.pin:
            amt = int(input("enter the amount that you want to deposit"))
            self.balance += amt
            if self.balance >= 500:
                self.transcations.append(f"Deposite : {amt}")
                print("Money has been deposited successfully", amt)
                self.menu()
            else:
                print("you have to deposite grater than 500")
        else:
            print("you have entered the wrong pin")
            self.deposit_money()

    def change_pin(self):
        user_pin = input("enter the pin to change your existing pin")
        if user_pin == self.pin:
            new_pin = input("enter the new pin")
            self.pin = new_pin
            print("Pin has been Changed successfully")
            self.menu()
        else:
            print("you have entered the wrong pin")
            self.menu()

    def withdraw_money(self):  # min balance 1000
        user_pin = input("enter the pin to withdraw the money")
        if user_pin == self.pin:
            amt = int(input("enter the amount you want to withdraw"))
            if self.balance > amt:
                rem = self.balance - amt
                print("Money has been withdrawed successfully", amt)
                if amt <= 1000:
                    charge = 0
                    self.balance = rem - charge
                    self.transcations.append(f"Withdrawn ₹{amt} | Charge ₹{charge}")
                    print("₹0 charge applied due to low balance")
                    self.menu()
                elif amt <= 20000:
                    charge = 100
                    self.balance = rem - charge
                    self.transcations.append(f"Withdrawn ₹{amt} | Charge ₹{charge}")
                    print("₹100 charge applied due to low balance")
                    self.menu()
                elif amt <= 100000:
                    charge = 1000
                    self.balance = rem - charge
                    self.transcations.append(f"Withdrawn ₹{amt} | Charge ₹{charge}")
                    print("₹1000 charge applied due to low balance")
                    self.menu()
                else:
                    charge = 2000
                    self.balance = rem - charge
                    self.transcations.append(f"Withdrawn ₹{amt} | Charge ₹{charge}")
                    print("₹2000 charge applied due to low balance")
                    self.menu()
            else:
                print("You have insufficient balance")
                self.menu()
        else:
            print("you have entered the wrong pin")
            self.menu()



    def check_balance(self):
        user_pin = input("enter your pin to check your balance")
        if user_pin == self.pin:
            print(f"Available Balance: {self.balance}")
            self.menu()
        else:
            print("the pin is invalid")
            self.menu()

    def transaction_history(self):
        user_pin = input("enter your pin to check transaction history")
        if user_pin == self.pin:
            print("-- Transcation History --")
            for i in self.transcations[-10:]:
                print(i)
            self.menu()
            
a = atm()
```
